```python
# Step 4: Transforming the "Education" Categorical Variable using an Ordinal Transformation
# First step is to change the data type into "Categorical"
df.Education= df.Education.astype('category')

#Second step is to provide the correct order of categories
df.Education= df.Education.cat.reorder_categories(['1. < HS Grad','2. HS Grad','3. Some College','4. College Grad','5. Advanced Degree'])

#Third step is to encode categories with integer values
df.Education= df.Education.cat.codes
df
```
